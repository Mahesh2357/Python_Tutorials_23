```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([3, 89])
y = np.array([2, 6])
plt.plot(x, y)
plt.xlabel ('X-axis')
plt.ylabel ('Y-axis')
plt.title('Line Graph')
plt.show()
```

```python
activity = ['eat', 'sleep', 'work', 'play']
slices = [3, 7, 6, 5]

plt.pie(slices, labels = activity, explode=(0, 0, 0.1, 0))    #####

plt.title('Pie Chart')
plt.show()
```

**How to display % using autopct**

```python
activity = ['eat', 'sleep', 'work', 'play']
slices = [3, 7, 6, 5]

plt.pie(slices, labels = activity, explode=(0, 0, 0.1, 0), autopct = '%1.1f%%')

plt.title('Pie Chart')
plt.show()
```

```python
activity = ['eat', 'sleep', 'work', 'play']
slices = [3, 7, 6, 5]

plt.pie(slices, labels = activity, explode=(0, 0, 0.1, 0), autopct = '%1.1f%%')

plt.title('Pie Chart')
plt.legend()                     ###
plt.show()
```
